# Back-translation

```python
# packages
!pip install sentencepiece
from transformers import M2M100ForConditionalGeneration, M2M100Tokenizer
```

```python
# Model and Tokenizer
model = M2M100ForConditionalGeneration.from_pretrained("facebook/m2m100_418M")
tokenizer = M2M100Tokenizer.from_pretrained("facebook/m2m100_418M")
```

```python
# from original to target language
original_ch = '硬漢是沒在過生日的'
original_encoded = tokenizer(original_ch, return_tensors="pt")
generated_tokens = model.generate(**original_encoded, forced_bos_token_id=tokenizer.get_lang_id("en"))
target_en = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
target_en
```

```python
# from target back to original language
target_encoding = tokenizer(target_en, return_tensors="pt", padding=True)
generated_tokens = model.generate(**target_encoding, forced_bos_token_id=tokenizer.get_lang_id("zh"))
```

```python
tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
```

# Synonyms

```python
!pip install nlpaug
```

```python
# packages
import nlpaug.augmenter.word as naw
```

```python
text = "The quick brown fox jumps over the lazy dog."
aug = naw.SynonymAug(aug_src='wordnet')
augmented_text = aug.augment(text)
print("Original:")
print(text)
print("Augmented Text:")
print(augmented_text)
```

# Random Crop

```python
text = "The quick brown fox jumps over the lazy dog."
aug = naw.RandomWordAug(action="crop")
augmented_text = aug.augment(text)
print("Original:")
print(text)
print("Augmented Text:")
print(augmented_text)
```
